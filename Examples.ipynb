# Layer Examples

```python
import os

import Layer
import numpy as np
```

```python
access_token = os.getenv('MAPBOX_ACCESS_TOKEN')
view_options = {
    'center': [37.17407153, 126.9086032],
    'zoom': 10,
    'bearing': 0,
    'pitch': 60,
    'frame_height': '700px',
    'style': 'mapbox://styles/mapbox/dark-v9',
    'access_token': access_token,
}
```

## Arc Layer

```python
data = [
    {
        'from': {
            'name': 'Hwaseong',
            'coordinates': [126.9086032, 37.17407153]
        },
        'to': {
            'name': 'Seoul',
            'coordinates': [126.986, 37.541]
        }
    }
]
```

```python
arc_layer = Layer.ArcLayer(data, **view_options)
```

```python
arc_layer.show()
```

## Path Layer

```python
data = [
    {
        'path': [[126.986, 37.541], [126.9086032, 37.17407153]],
        'name': 'Seoul-Hwaseong',
        'color': [255, 30, 100]
    },
    {
        'path': [[126.986, 37.541], [127.179108, 37.245635]],
        'name': 'Seoul-Suwon',
        'color': [200, 100, 30]
    }
]
```

```python
path_layer = Layer.PathLayer(data, 
                             width=5,
                             widthScale=20,
                             widthMinPixels=2,
                             **view_options)
```

```python
path_layer.show()
```

```python
trips = []
time_unit = 1
for data in paths:
    segments = []
    for idx, point in enumerate(data['path']):
        segments.append([point[0], point[1], idx*time_unit])
    data['segments'] = segments
    del data['path']
    trips.append(data)
```

```python
trip_layer = Layer.TripsLayer(trips, 
                             width=5,
                             widthScale=20,
                             widthMinPixels=2,
                             currentTime=10,
                             rounded=True,
                             **view_options)
trip_layer.show()
```

## Overay Layers

```python
data = [
    {
        'from': {
            'name': 'Hwaseong',
            'coordinates': [126.9086032, 37.17407153]
        },
        'to': {
            'name': 'Seoul',
            'coordinates': [126.986, 37.541]
        }
    }
]

data2 = [
    {
        'from': {
            'name': '경주',
            'coordinates': [128.162384, 36.416862]
        },
        'to': {
            'name': 'Seoul',
            'coordinates': [126.986, 37.541]
        }
    }
]
```

```python
m = Layer.Map(**view_options)
```

```python
arc_layer = Layer.ArcLayer(data, **view_options)
arc_layer2 = Layer.ArcLayer(data2, **view_options)
m.add(arc_layer)
m.add(arc_layer2)
```

```python
m.show()
```
